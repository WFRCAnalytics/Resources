```python
import os
import pandas as pd
from dbfread import DBF
import sqlite3
```

```python
dbfNode = r"E:\GitHub\WF-TDM-v9x\Scenarios\v910-official\BY_2019\0_InputProcessing\UpdatedMasterNet\GIS\WFv910_MasterNet - Node.DBF"

dfNodes = pd.DataFrame(DBF(dbfNode, load=True))
dfNodes
```

```python
#Make the db in memory
conn = sqlite3.connect(':memory:')
dfNodeNumberingScheme.to_sql('dfNodeNumberingScheme', conn, index=False)
dfNodes.to_sql('dfNodes', conn, index=False)

qry = '''
    select  
        *
    from
        dfNodes join dfNodeNumberingScheme on
        N between N_FROM and N_TO
    '''
dfNodesWithScheme = pd.read_sql_query(qry, conn)
display(dfNodesWithScheme)
```

```python
dfTransitNodes = dfNodesWithScheme[dfNodesWithScheme['N_GROUP']=='Transit']
dfTransitNodes
```

```python
# filter by those with Names
dfTransitNodesWithNames = dfTransitNodes[(dfTransitNodes['NODENAME'].str.strip() != "")]
dfTransitNodesWithNames = dfTransitNodesWithNames[['N','NODENAME','X','Y','GEOGKEY','TAZID']]

# remove station from names to be consistent
dfTransitNodesWithNames['NODENAME'] = dfTransitNodesWithNames['NODENAME'].str.replace(" Station", "").str.replace(" Bridge", "")

dfTransitNodesWithNames
```

```python
dfTransitNodesWithNames.to_csv('node-station-names.csv', index=False)
```
