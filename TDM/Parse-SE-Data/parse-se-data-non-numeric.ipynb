```python
import os
import glob
import re
import pandas as pd
```

```python
se_folder_path = r'E:\GitHub\WF-TDM-v9x\1_Inputs\2_SEData\2_WFRC\FiscallyConstrained'
```

```python
# clean out output folder

files = glob.glob(os.path.join('output', '*'))

for file in files:
    try:
        os.remove(file)
        print(f"Deleted: {file}")
    except Exception as e:
        print(f"Error deleting {file}: {e}")
```

```python
# get SE data
df_se_data = pd.DataFrame()

# Loop through the years 2019 to 2050
for se_year in range(2019, 2051):

    file = "SE_" + str(se_year) + '.csv'
    se_file = os.path.join(se_folder_path, file)

    _dbf = pd.read_csv(se_file)

    # Identify non-numeric columns
    non_numeric_columns = _dbf.select_dtypes(include=[int, float]).columns.tolist()
    non_numeric_columns.remove(';TAZID')  # Keep 'TAZID' as identifier variable

    _df_pivot = _dbf.melt(id_vars=[';TAZID'], value_vars=non_numeric_columns)

    _df_pivot['se_year'] = se_year

    df_se_data = pd.concat([df_se_data,_df_pivot])

df_se_data.rename(columns={';TAZID':'TAZID','variable':'parameter'}, inplace=True)

display(df_se_data)
```

```python
df_se_data.to_csv('output/se_data_long.csv', index=False)
```
