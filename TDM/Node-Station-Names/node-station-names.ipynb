```python
import os
import pandas as pd
from dbfread import DBF
import sqlite3
```

```python
dfNodeNumberingScheme = pd.read_csv(r'..\NodeNumberingScheme.csv')
display(dfNodeNumberingScheme)

dfTrn = dfNodeNumberingScheme[dfNodeNumberingScheme['N_GROUP']=='Transit']
dfTrn

```

```python
dbfNode = r"E:\GitHub\WF-TDM-v9.0\1_Inputs\3_Highway\GIS\MasterNet_v9 - 2023-03-02 - Node.DBF"

dfNodes = pd.DataFrame(DBF(dbfNode, load=True))
dfNodes
```

```python
#Make the db in memory
conn = sqlite3.connect(':memory:')
dfNodeNumberingScheme.to_sql('dfNodeNumberingScheme', conn, index=False)
dfNodes.to_sql('dfNodes', conn, index=False)

qry = '''
    select  
        *
    from
        dfNodes join dfNodeNumberingScheme on
        N between N_FROM and N_TO
    '''
dfNodesWithScheme = pd.read_sql_query(qry, conn)
display(dfNodesWithScheme)
```

```python
dfTransitNodes = dfNodesWithScheme[dfNodesWithScheme['N_GROUP']=='Transit']
dfTransitNodes
```

```python
# filter by those with Names
dfTransitNodesWithNames = dfTransitNodes[(dfTransitNodes['NODENAME'].str.strip() != "")]
dfTransitNodesWithNames = dfTransitNodesWithNames[['N','NODENAME','X','Y','GEOGKEY']]

# remove station from names to be consistent
dfTransitNodesWithNames['NODENAME'] = dfTransitNodesWithNames['NODENAME'].str.replace(" Station", "")

dfTransitNodesWithNames
```

```python
csvFR   = r"E:\GitHub\Transit-2019-Fall\data\UTA_Average_Weekday_Rail_Station_Boardings_Sep-Nov_2019_raw_FrontRunner.csv"
csvTRAX = r"E:\GitHub\Transit-2019-Fall\data\UTA_Average_Weekday_Rail_Station_Boardings_Sep-Nov_2019_raw_TRAX.csv"

dfFR   = pd.read_csv(csvFR  )
dfTRAX = pd.read_csv(csvTRAX)

dfTRAX.rename(columns={'FX_LineAbbr':'LINE','NodeName':'STATION'},inplace=True)
dfFR  .rename(columns={'station':'STATION'}                      ,inplace=True)
dfTRAX['STATION'] = dfTRAX['STATION'].str.replace(" Station", "")

#dfFR[['STATION']] = dfFR[['STATION']] + ' Station'

dfUTAStations = pd.concat([dfFR[['STATION']].drop_duplicates(),dfTRAX[['STATION']].drop_duplicates()]).drop_duplicates()

#display(dfFR  )
#display(dfTRAX)
display(dfUTAStations)
```

```python
dfUTAtoTDM = pd.DataFrame.merge(dfUTAStations, dfTransitNodesWithNames, left_on='STATION', right_on='NODENAME', how='outer')
dfUTAtoTDM
```

```python
dfOBS = pd.read_csv(r"E:\GitHub\Transit-2019-Fall\data\2019-OBS\2019 Final Weighted UTA OD Data - 2022-04-05 - processed.csv",low_memory=False)
dfOBS
```

```python
pd.set_option('display.max_columns', None)
dfOBS
```

```python
dfOBSRail = dfOBS[dfOBS['Mode_Fin'].isin([7,8,9])] # all LRT, CRT, BRT
dfStopOn  = dfOBSRail[['stop_on_clntid' ,'stop_on_addr' ,'stop_on_lat' ,'stop_on_long' ]].rename(columns={'stop_on_clntid' :'STATION_OBS_ID','stop_on_addr' :'STATION_OBS','stop_on_lat' :'LAT','stop_on_long' :'LON'}).dropna().drop_duplicates()
dfStopOff = dfOBSRail[['stop_off_clntid','stop_off_addr','stop_off_lat','stop_off_long']].rename(columns={'stop_off_clntid':'STATION_OBS_ID','stop_off_addr':'STATION_OBS','stop_off_lat':'LAT','stop_off_long':'LON'}).dropna().drop_duplicates()

dfOBSStops = pd.concat([dfStopOn,dfStopOff]).drop_duplicates()

dfOBSStops['STATION'] = dfOBSStops['STATION_OBS'].apply(str.title).str.replace(" Station", "")
display(dfOBSStops)

#dfStops
```

```python
dfUTAtoTDMtoOBS = pd.DataFrame.merge(dfUTAtoTDM, dfOBSStops, on='STATION', how='outer')
dfUTAtoTDMtoOBS
```
